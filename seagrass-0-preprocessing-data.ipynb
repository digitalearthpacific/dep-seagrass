```python
import geopandas as gpd
import pandas as pd
```

```python
yanuca_east = gpd.read_file("seagrass/yanuca_east_1.geojson")
yanuca_west = gpd.read_file("seagrass/yanuca_west_1.geojson")
leluvia = gpd.read_file("seagrass/leluvia_1.geojson")
```

```python
# Preprocess Leluvia
leluvia=leluvia.replace("Seagrass", "seagrass")
leluvia=leluvia.replace("Sand", "sand")
leluvia=leluvia.replace("Coral", "coral")
leluvia=leluvia.replace("Rock", "rock")
leluvia=leluvia.replace("Notes", "lulc")
leluvia = leluvia.rename(columns={'Notes': 'lulc'})
```
